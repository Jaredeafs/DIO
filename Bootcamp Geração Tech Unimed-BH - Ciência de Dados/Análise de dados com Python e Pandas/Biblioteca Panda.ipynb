## **Introdução a Biblioteca Pandas**

```python
#Importando a biblioteca pandas
import pandas as pd       #Aqui estamos importando a biblioteca e dando o apelido de pd para ela
                          #Toda vez que formos chamar essa biblioteca novamente só precisaremos escrever pd
```

```python
 #lei um arquivo csv

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Gapminder.csv", sep=";")
```

```python
#Visualizando as 5 primeiras linhas

df.head()
```

```python
#Renomeando titulos para portugues

df = df.rename(columns={"country":"Pais", "continent": "continente", "year":"Ano", "lifeExp":"Expectativa de vida", "pop":"Pop Total", "gdpPercap":"PIB"})
```

```python
df.head()
```

```python
#Para retornar apenas o nome das colunas

df.columns
```

```python
# Para saber o tipo de dado em cada coluna

df.dtypes
```

```python
# Mostrar as ultimas 5 linhas 
#obs.: para apresentar numeros de linhas determinadas basta estabelecer o numero
#      de linhas desejadas dentro dos parenteses como parametro

df.tail()
```

```python
# Para retornar apenas os continentes que temos em nossa base

df["continente"].unique()
```

```python
#Para retornar apenas os valos de um continente expecifico
#Vamos de exemplo com Oceania

Oceania = df.loc[df["continente"] == "Oceania"]     #comando loc para filtrar
Oceania.head()
```

```python
Oceania["continente"].unique()
```

```python
# Podemos tambem fazer agrupamentos de dados usando o groupby
# Exemplo: digamos que queremos agrupar por continente e dizer quantos paises temos por continente

df.groupby("continente")["Pais"].nunique()
```

```python
# Agora eu quero saber a expectativa de vida media para cada ano em nossos dados

df.groupby("Ano")["Expectativa de vida"].mean()        #comando mean para media
```

```python
# Vamos fazer com a media do PIB

df["PIB"].mean()
```

```python
# vamos retornar a soma do PIB

df["PIB"].sum()
```

## **1. Listas**



```python
#Criando uma lista chamada animais
animais = [1,2,3]#aqui temos uma lista apenas com tipos inteiros
animais
```

```python
animais = ["cachorro", "gato", 12345, 6.5] #aqui temos uma lista com varios tipos
animais
animais[0] #aqui estamos imprimindo o primeiro elemento da lista
```

```python
#Imprimindo o 4 elemento da lista
animais [3]
```

```python
#Subistituindo o primerio elemento da lista
animais[0] = "papagaio"
animais
```

```python
#Removendo gato da lista
animais.remove("gato")
animais
```

```python
#Vamos descobrir o tamanho da lista
len(animais)
```

```python
#Como saber que um elemento pertence aquela lista
"gato" in animais
```

```python
#Vamos criar uma nova lista

new_list = [500, 30, 300, 80, 10]
new_list
```

```python
max(new_list)
```

```python
min(new_list)
```

```python
#como adicionar um elemento a nossa lista
animais.append("leão")
animais
```

```python
animais.extend(["cobra", 6])
animais
```

```python
#Também podemos usar o metodo count para saber quantas vezes o elemento aparece em nossa lista
animais.count("leão")
```

## **4. DICIONARIOS**



```python
#Para criar um dicionário utilizamos as {}
dc = {"Maçã":20, "Banana":10 , "Laranja":15, "Uva":5} 
                  #Dicionários trabalham com o conceito chave e valor
dc
```

```python
#Como acessar o valor de um dicionario atraves da chave
dc["Maçã"]
```

```python
#Atualizando o valor da Maçã
dc["Maça"] = 25
dc
```

```python
#Retornando todas as chaves do dicionario
dc.keys()
```

```python
#Retornando os valores contindos em cada chave do dicionario
dc.values()
```

```python
#Verificando se já existe uma chave no dicionário e caso não exista inserir ela
dc.setdefault("Limão",22)
```

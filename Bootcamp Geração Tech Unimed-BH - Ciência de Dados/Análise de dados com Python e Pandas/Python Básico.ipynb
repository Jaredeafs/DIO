```python
a= 12
b= 13

r = a + b

print(r)
```

```python
# soma
r_soma = a+b

# subtração
r_subtração = a-b

# divisão
r_divisão = a/b

# multiplicação
r_multiplicação = a*b

print("Soma =          {}".format(r_soma))
print("Subtração =     {}".format(r_subtração))
print("Divisão =       {:.2f}".format(r_divisão))
print("Multiplicação = {}".format(r_multiplicação))
```
